```python
df = pd.DataFrame(([s.date, s.open, s.close, s.asset.ticker, s.interday_volatility, s.volume] for s in dh.StockRowIterator(stocks)),
                  columns=['date', 'open', 'close', 'asset', 'return', 'volume'])
```

```python
ftse = df.loc[df['asset']=="FTSE"]
ftmc = df.loc[df['asset']=="^FTMC"]
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline 

import matplotlib
matplotlib.rc('xtick', labelsize=40) 
matplotlib.rc('ytick', labelsize=40) 

import seaborn as sns
sns.set(style="whitegrid", color_codes=True)
from statsmodels.tsa.stattools import acf, pacf

df2 = ftse[['date','return']].sort_values('date').set_index('date')

df3 = ftmc[['date', 'return']].sort_values('date').set_index('date')

# plt.figure()
# plt.plot(df2, marker=".")
# plt.plot(df3, marker=".")
# plt.show()

# lag_acf1 = plot_acf(df2, lags=10, fft=True)
# lag_pacf1 = plot_pacf(df2, lags=10)

# lag_acf2 = plot_acf(df3, lags=10, fft=True)
# lag_pacf2 = plot_pacf(df3, lags=10)
```

```python
from statsmodels.tsa.stattools import adfuller
import hurst



def split(df):
    index = round(len(df)*0.8)
    train = df.iloc[:index]
    test = df.iloc[index:]
    return train, test

def adf_stat(df):
    result = adfuller(df)
    # low adf statistic tells us that the series is non stationary
    print('ADF Statistic: %f' % result[0])
    # high p value (>=0.05) also tells us the series is non stationary
    print('p-value: %f' % result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))
    return result

def hurst_stat(df):
    H,c,data = hurst.compute_Hc(df)
    # H < 0.5 is stationary 
    # H > 0.5 is non-stationary or trending
    # H = 0.5 is random walk/Brownian motion
    print("H = {:.4f}, c = {:.4f}".format(H,c))


    
train1, test1 = split(df2)
train2, test2 = split(df3)

adf_stat(train1)
adf_stat(train2)

hurst_stat(train1)
hurst_stat(train2)
```

```python
from statsmodels.tsa.arima_model import ARIMA
from math import sqrt
from sklearn import metrics

model = ARIMA(train1.values, order=(2, 0, 1))
model_fit = model.fit(disp=False)

model2 = ARIMA(train2.values, order=(1, 0, 1))
model2_fit = model2.fit(disp=False)
predictions = model_fit.predict(len(test1))
test_ = test1.copy()
test_['predictions'] = predictions[0:len(test1)]

plt.plot(df2)
plt.plot(test_.predictions)
plt.show()

error = sqrt(metrics.mean_squared_error(test1.values,predictions[0:len(test1)]))
print('Test RMSE for ARIMA: ', error)
```
